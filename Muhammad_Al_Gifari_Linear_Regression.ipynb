```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv('/content/Mall_Customers.csv')
df
```

```python
#menentukan siapa yang jadi input x dan target y
X= df[['Age']]
y= df['Spending Score (1-100)']
```

```python
#regresi linear
model= LinearRegression()
model.fit(X,y)
```

```python
plt.scatter(X, y, color='blue', alpha=0.5 , label='Data Pelanggan Asli')
plt.plot(X, model.predict(X), color='red', linewidth=3, label='Garis Tebakan (Regresi)')

plt.title('Menebak skor pengeluaran pelanggan berdasarkan umur')
plt.xlabel('Umur')
plt.ylabel('Skor Pengeluaran')
plt.legend()
plt.show()

```

"Berdasarkan pemodelan Regresi Linear pada dataset pelanggan ritel, ditemukan korelasi negatif yang signifikan antara usia dan skor pengeluaran, yang berarti daya beli pelanggan cenderung menurun seiring bertambahnya usia. Analisis menunjukkan bahwa kelompok usia produktif (20-40 tahun) merupakan motor penggerak utama pendapatan toko dengan skor pengeluaran menyentuh angka maksimal (100), sementara segmen senior (50-70 tahun) menahan pengeluarannya pada batas skor 60. Oleh karena itu, strategi bisnis yang direkomendasikan adalah memfokuskan alokasi budget marketing pada demografi muda untuk memaksimalkan Return on Investment (ROI), sembari mengevaluasi ketersediaan variasi produk yang relevan dengan kebutuhan kesehatan atau gaya hidup untuk memancing transaksi dari segmen usia lanjut."
